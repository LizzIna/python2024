```python
import matplotlib.pyplot as plt

plt.plot(vector, label = "Vector")
plt.title("Grafico")
plt.xlabel("indice")
plt.ylabel("valores")

plt.grid(True)#agrega cuadricula
plt.legend()#Invoca propiedad label de plot
plt.show()#Visualiza el grafico con python
```
